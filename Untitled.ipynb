## Molding data for deep learning

```python
del df2["instant"]
del df2["casual"]
del df2["registered"]
del df2["temp"]
del df2["season"]
```

```python
num_obs = len(df2)
num_first_year = len(df2.loc[df2["yr"] == 0])
num_second_year = len(df2.loc[df2["yr"] == 1])
print("Number of instances in the first year:  {0} ({1:2.2f}%)".format(num_first_year, (num_first_year/num_obs) * 100))
print("Number of instances in the second year: {0} ({1:2.2f}%)".format(num_second_year, (num_second_year/num_obs) * 100))
```

```python
from sklearn.model_selection import train_test_split
feature_col_names = ["yr", "mnth", "holiday", "weekday", "workingday", "weathersit", "atemp", "hum", "windspeed"]
predicted_class_names = ["cnt"]

X = df2[feature_col_names].values     # predictor feature columns 
y = df2[predicted_class_names].values # predicted class (1=true, 0=false) column

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=40)
```

```python
print("Original first year  : {0} ({1:0.2f}%)".format(len(df.loc[df["yr"] == 0]), (len(df.loc[df["yr"] == 0])/len(df.index)) * 100.0))
print("Original second year : {0} ({1:0.2f}%)".format(len(df.loc[df["yr"] == 1]), (len(df.loc[df["yr"] == 1])/len(df.index)) * 100.0))
print("")
print("Training first year  : {0} ({1:0.2f}%)".format(len(X_train[X_train[:, 0] == 0]), (len(X_train[X_train[:, 0] == 0])/len(X_train) * 100.0)))
print("Training second year : {0} ({1:0.2f}%)".format(len(X_train[X_train[:, 0] == 1]), (len(X_train[X_train[:, 0] == 1])/len(X_train) * 100.0)))
print("")
print("Test first year      : {0} ({1:0.2f}%)".format(len(X_test[X_test[:, 0] == 0]), (len(X_test[X_test[:, 0] == 0])/len(X_test) * 100.0)))
print("Test second year     : {0} ({1:0.2f}%)".format(len(X_test[X_test[:, 0] == 1]), (len(X_test[X_test[:, 0] == 1])/len(X_test) * 100.0)))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Building a model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(9, activation="relu"))
model.add(Dense(20, activation="sigmoid"))
model.add(Dense(20, activation="relu"))
model.add(Dense(10, activation="relu"))


# Binary Classification
model.add(Dense(1, activation="relu"))

model.compile(loss="mse", optimizer="adam")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20)
```

```python
model.fit(x=X_train, y=y_train, epochs=400, validation_data=(X_test, y_test),
          callbacks=[early_stop], batch_size=64, verbose=1)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

## Model evaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
mean_absolute_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions)
```

```python
# Our predictions
plt.scatter(y_test,predictions)

# Perfect predictions
plt.plot(y_test,y_test,'r')
```
