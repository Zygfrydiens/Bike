```python
num_obs = len(df)
num_first_year = len(df.loc[df["yr"] == 0])
num_second_year = len(df.loc[df["yr"] == 1])
print("Number of instances in the first year:  {0} ({1:2.2f}%)".format(num_first_year, (num_first_year/num_obs) * 100))
print("Number of instances in the second year: {0} ({1:2.2f}%)".format(num_second_year, (num_second_year/num_obs) * 100))
```

```python
from sklearn.model_selection import train_test_split
feature_col_names = ["yr", "mnth", "hr", "holiday", "weekday", "workingday", "weathersit", "atemp", "hum", "windspeed"]
predicted_class_names = ["cnt"]

X = df[feature_col_names].values     # predictor feature columns 
y = df[predicted_class_names].values # predicted class (1=true, 0=false) column

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=40)
```

```python
print("Original first year  : {0} ({1:0.2f}%)".format(len(df.loc[df["yr"] == 0]), (len(df.loc[df["yr"] == 0])/len(df.index)) * 100.0))
print("Original second year : {0} ({1:0.2f}%)".format(len(df.loc[df["yr"] == 1]), (len(df.loc[df["yr"] == 1])/len(df.index)) * 100.0))
print("")
print("Training first year  : {0} ({1:0.2f}%)".format(len(X_train[X_train[:, 0] == 0]), (len(X_train[X_train[:, 0] == 0])/len(X_train) * 100.0)))
print("Training second year : {0} ({1:0.2f}%)".format(len(X_train[X_train[:, 0] == 1]), (len(X_train[X_train[:, 0] == 1])/len(X_train) * 100.0)))
print("")
print("Test first year      : {0} ({1:0.2f}%)".format(len(X_test[X_test[:, 0] == 0]), (len(X_test[X_test[:, 0] == 0])/len(X_test) * 100.0)))
print("Test second year     : {0} ({1:0.2f}%)".format(len(X_test[X_test[:, 0] == 1]), (len(X_test[X_test[:, 0] == 1])/len(X_test) * 100.0)))
```
